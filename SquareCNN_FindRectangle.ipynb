```python
def fit_rotated_rectangle(image_path):
    # Read the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply edge detection (e.g., Canny edge detection) to detect edges in the image
    edges = cv2.Canny(image, 100, 200)

    # Apply the Hough Line Transform to detect straight lines in the edge-detected image
    lines = cv2.HoughLines(edges, rho=1, theta=np.pi/180, threshold=100)

    # Find the four most prominent lines that form the sides of the rectangle
    sorted_lines = sorted(lines[:, 0], key=lambda x: x[1], reverse=True)[:4]

    # Calculate the intersection points of the four lines to get the rectangle vertices
    vertices = []
    for i in range(4):
        for j in range(i+1, 4):
            rho1, theta1 = sorted_lines[i]
            rho2, theta2 = sorted_lines[j]
            A = np.array([[np.cos(theta1), np.sin(theta1)], [np.cos(theta2), np.sin(theta2)]])
            b = np.array([rho1, rho2])
            intersection = np.linalg.solve(A, b)
            intersection = np.round(intersection).astype(int)
            vertices.append(intersection)

    # Draw the fitted rectangle on the original image
    for i in range(4):
        cv2.line(image, tuple(vertices[i]), tuple(vertices[(i+1)%4]), 255, 2)

    # Display the image with the fitted rectangle
    cv2.imshow("Fitted Rectangle", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage:
image_path = "path_to_your_image.png"
fit_rotated_rectangle(image_path)
```
