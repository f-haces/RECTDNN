```python
import os
import shutil
import re
```

```python
def crawl_directory(source_dir, destination_dir):
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            if file.lower().endswith(('.jpg', '.tif', '.png')):
                image_path = os.path.join(root, file)
                world_file = get_world_file(image_path)
                if not world_file:
                    move_file(image_path, destination_dir)

def get_world_file(image_path):
    base_path, ext = os.path.splitext(image_path)
    world_extensions = ['.jpw', '.tfw', '.pgw', '.pnw']
    for world_ext in world_extensions:
        world_file = base_path + world_ext
        if os.path.exists(world_file):
            return world_file
    return None

def move_file(image_path, destination_dir):
    file_name = os.path.basename(image_path)
    destination_path = os.path.join(destination_dir, file_name)
    shutil.move(image_path, destination_path)
    print(f"Moved {file_name} to {destination_dir}")

# Usage example
source_directory = r'C:\Users\fhacesga\Desktop\FIRMsDigitizing\RECTDNN\data\000_WorkingFiles\Known'
destination_directory = r'C:\Users\fhacesga\Desktop\FIRMsDigitizing\RECTDNN\data\000_WorkingFiles\Unknown'

crawl_directory(source_directory, destination_directory)
```
