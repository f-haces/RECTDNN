```python
import os
import shutil
import re
```

```python
directory_a = r"D:\FloodChange\WorkingDatasets\Current\\"
directory_b = r"D:\FloodChange\WorkingDatasets\History\\"
```

```python
# Get the list of image files in directory B
reference_images_fullnames = [file for file in os.listdir(directory_a) if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.tif'))]
```

```python
end_dict = {
    ".jpg" : ".jpw",
    ".tif" : ".tfw",
    ".png" : ".pnw"
}
```

```python
for ref_image in reference_images_fullnames:
    full_image_dir = directory_a + ref_image
    world_file_dir = directory_a + ref_image.split('.')[0] + end_dict['.'+ref_image.split('.')[1]]
    
    print(full_image_dir)
    print(world_file_dir)
    if not os.path.isfile(world_file_dir):
        print("deleting")
        os.remove(full_image_dir)
```

```python
image_files_b = [file for file in os.listdir(directory_b) if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.tif'))]
reference_images_fullnames = [file for file in os.listdir(directory_a) if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.tif'))]
reference_images_ids = [''.join(re.findall(r'\d+', file)) for file in reference_images_fullnames]
```

```python
image_files_b
```

```python
for image_file_b in image_files_b:
    print(image_file_b, end=" ")
    try:
        # Extract the numerical portion of the image file's name
        base_name_b = os.path.splitext(image_file_b)[0]
        extension = image_file_b[-4:]
        numerical_part_b = ''.join(re.findall(r'\d+', base_name_b))

        try:
            currindex = reference_images_ids.index(numerical_part_b)
        except ValueError:
            currindex = -1
        infile_name  = directory_a + reference_images_fullnames[currindex].split('.')[0] + end_dict[extension]
        outfile_name = directory_b + base_name_b + end_dict[extension]
        if not os.path.isfile(outfile_name) and currindex != -1:
            shutil.copyfile(infile_name, outfile_name)
            print("... copied")
        else:
            print("... other file")
    except:
        print("... error")
        continue
```
